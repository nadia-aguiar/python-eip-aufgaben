# Aufgabe 1
## a
Ein **Algorithmus** ist ein Ablauf, dessen Schritte klar definiert sind und die nicht verändert werden und welcher mehrmals wiederholt werden kann. Die einzelnen Schritte sind so klar definiert, dass sie theoretisch auch von einem Computer ausgeführt werden könnten. Ein Algorithmus dient meist der Lösung eines sehr spezifischen Problems.
Ein Algorithmus ist programmiersprachenunabhängig und kann auch in normaler Umgangssprache (oder einem Diagramm) formuliert werden und kann auch als eine Art Anleitung für Menschen geschrieben sein. Eine Anleitung zum Zusammenbau eines Regals könnte beispielsweise als ein niedergeschriebener Algorithmus interpretiert werden, wenn man unbedingt wollte.

Ein **Programm** ist ein Algorithmus, der für Computer geschrieben wird und von ihnen ausgeführt werden kann und soll. Es muss entweder in Maschinensprache geschrieben sein, oder in einer Sprache, die sich in Maschinensprache übersetzen lässt, "Programmiersprache" genannt. Normale Umgangssprache ist hier nicht möglich.
Eine weitere Definition bezeichnet ein Programm als die Realisierung eines Algorithmus (implizit wohl durch eine Programmiersprache). Also sinngemäß "Ein Algorithmus ist der abstrahierte Ablauf; ein Programm ist, wenn man diesen tatsächlich mittels einer Programmiersprache hinschreibt, kompilieren/interpretieren und ausführen lässt".

## b
* Ein **Compiler** übersetzt Quellcode in einer Programmiersprache/"Hochsprache" in Maschinencode, der vom Computer ausgeführt werden kann.
* Ein **Interpreter** erzeugt keinen Maschinencode; sondern zum Beispiel Bytecode für eine hypothetische virtuelle CPU, wobei dann während der Ausführung der aktuelle Bytecode-Teil in Maschinencode übersetzt wird. Der Quellcode wird scheinbar direkt ausgeführt.
* Ein **statisches Typsystem** bestimmt Datentypen zur Compilezeit und ermöglichen das Finden von entsprechenden Fehlern schon zur Compilezeit und damit, bevor das Programm überhaupt ausgeführt wird. Dies bedeutet auch, dass der Programmierer im Vergleich zu dynamischen Typsystemen mehr Informationen angeben muss und evtl Typkonvertierungen vornehmen muss.
* Ein **dynamisches Typsystem** bestimmt Datentypen erst zur Laufzeit und der Programmierer muss weniger Informationen angeben, da das Programm, umgangssprachlich formuliert, "erst mal versucht, selbst heraus zu finden, was der Programmierer wahrscheinlich gemeint hat" - was aber auch fehleranfällig ist; weil das Programm evtl etwas anderes als "gemeint" interpretiert, als tatsächlich intendiert. Und viele Fehler treten erst zur Laufzeit auf.
* **Compilezeit** bezeichnet vermutlich die Zeit, in der das Programm kompiliert wird; scheint aber auch mit den Prozessen zu tun zu haben, die während der Kompilierung laufen (wie beispielsweise lexikalische und syntaktische Analyse und Typechecking)
* **Laufzeit** bezeichnet hier in dem Kontext die Zeit, in der das Programm ausgeführt wird und aktiv ist.

# Aufgabe 4
## Skizze des Ausführungsmodells
Zunächst wird in einem Editor der sogenannte Quellcode geschrieben. Dieser besteht aus de eigentlichen Code und Kommentaren, die zum besseren Verständnis eingefügt werden. 

Der Quellcode direkt kann nicht von einem Computer ausgeführt werden, sondern muss dafür in eine andere Form "übersetzt" werden, den sogenannten Bytecode. Das macht der Compiler in mehreren Einzelschritten. 

Im ersten Schritt führt der Compiler eine lexikalische Analyse durch. Der Lexer zerlegt den Code in einzelne Bestandteile (sog. Lexeme) und weißt ihnen dann Token zu. Die Token sind grammatikalische Kategorien. Nun haben wir eine Abfolge von Lexemen und Token.

Danach folgt die syntaktische Analyse, durchgeführt vom Parser. 

Dieser führt eine Überprüfung der Grammatik durch und schaut ob diese in Python erlaubt ist. Dabei erzeugt er einen Parsebaum oder Abstract Syntax Tree (AST), der für die Interpretation benötigt wird.

Der Interpreter kann diesen Bytecode nun ausführen und auf dem Computer lässt sich eine Ausgabe entnehmen.

Bei der Entwicklung neuer Software würde man nun überprüfen, ob man die gewünschte Ausgabe erhält. Wenn nicht müsste man wieder am Quellcode ansetzen und ihn verbessern, bis man die gewünschte Ausgabe erhält.


## Fallbeispiele
* Vergessene schließende Klammer im Quellcode: Compiler kann den Code nicht übersetzten, da der Lexer zwar alles zuordnen kann, aber dem Parser die fehlende Klammer "auffällt"
* Division zweier Ganzzahlen mit Fließkommazahl als Ergebnis wird als Ganzzahl behandelt: Der Fehler kommt erst bei der Ausführung/Ausgabe zum Tragen, da ja der Compiler kein Problem damit hat, dass der Quotient zweier Ganzzahlen auch eine Ganzzahl (und evtl nicht präzise) ist - sondern Menschen; denn das Ergebnis wird  dafür gerundet und evtl nicht ganz korrekt.
* Nutzereingabe einer Zahl, die positiv sein muss, es wird aber -2 eingegeben: eine negative Zahl ist immer noch vom Typ int, von daher werden vermutlich weder Lexer noch Parser noch Compiler/Interpreter damit Probleme haben. Das Programm tut aber evtl unerwartete Dinge, die man als Mensch nicht möchte. Also wahrscheinlich kommt auch so ein Fehler erst bei der Ausführung/Ausgabe zum Tragen.
