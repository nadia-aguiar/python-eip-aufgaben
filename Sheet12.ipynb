# Aufgabe 1

Def: $R_n := R_{n-1} +2^n$

Def: $R_0 := 1$



**Behauptung:** Es gilt $R_n = 2^{n+1} -1$ für alle $n \in \mathbb{N}$.

**Beweis:**

*IA:*

$n = 0\\
R_0 = 2^{0+1} - 1 = 2^1 - 1 = 2 - 1 = 1$



*IV:* Es gelte $R_n = 2^{n+1} - 1$ für ein festes $n \in \mathbb{N}$

*IS:*

zu zeigen: $R_{n+1} = 2^{n+2} - 1$
\begin{align*}
R_{n+1} &\overset{\text{def}}{=} R_n + 2^{n+1} \\
&\overset{\text{IV}}{=}  2^{n+1} - 1 + 2^{n+1} \\
&= 2^{n+1} + 2^{n+1} - 1 \\
&= 2\cdot 2^{n+1} - 1 \\
&= 2^{n+2} - 1 \\
\end{align*}

```python
import math
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

def linear(n):
    linear_list = []
    for n in range (2,13):
        linear_list.append (n)
        n += 1
    return linear_list

def squared_two(n):
    squared_two_list = []
    for n in range (2,13):
        squared_two_list.append (n**2)
        n += 1
    return squared_two_list
        
def squared_three(n):
    squared_three_list = []
    for n in range (2,13):
        squared_three_list.append (n**3)
        n += 1
    return squared_three_list
    
def squared_n(n):
    squared_n_list = []
    for n in range (2,13):
        squared_n_list.append (2**n)
        n += 1
    return squared_n_list
        
def n_log_n(n):
    n_log_list = []
    for n in range (2,13):
        n_log_list.append (n * math.log(n,10))
        n += 1
    return n_log_list


def log_n(n):
    log_list = []
    for n in range (2,13):
        log_list.append(math.log(n,10))
        n += 1
    return log_list


x = list(range(2, 13))

#Funktionen plotten
plt.plot(x, linear(2), label = 'linear')
plt.plot(x, squared_two(2), label = 'n^2')
plt.plot(x, squared_three(2), label = 'n^3')
plt.plot(x, squared_n(2), label = '2^n')
plt.plot(x, n_log_n(2), label = 'n*log(n)')
plt.plot(x, log_n(2), label = 'log(n)')

#y-Achse begrenzen
plt.ylim([0, 175])

#Achsenbeschriftung
plt.xlabel('n')
plt.ylabel('f(n)')

#Legende
plt.legend()

#speichern
plt.savefig("linear_scale.png")
```
