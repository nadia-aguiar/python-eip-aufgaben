```python
# Eingabe der Zahlen durch den Benutzer
x = int(input("Bitte geben Sie die erste Zahl ein!"))
y = int(input("Bitte geben Sie die zweite Zahl ein!"))
# Initialisiere k mit Null 
k = 0
#Solange keine der Zahlen Null erreicht hat
while(not((x == 0) or (y == 0))):
    if ((x % 2)  == 0):
        if ((y % 2) == 0): #Fall: beide Zahlen sind durch 2 teilbar
            x = x / 2
            y = y / 2
            k = k + 1
        else: # Fall: nur x ist durch 2 teilbar
            x = x / 2 
    else:
        if ((y % 2) == 0): #Fall: nur y ist durch 2 teilbar
            y = y / 2
        else: #Fall: keine der Zahlen ist durch 2 teilbar
            if x > y:
                x = (x - y) / 2
            else:
                y = (y - x) / 2  
#Da die Schleife beendet ist, ist mindestens eine der Zahlen gleich Null. Berechnung ggT                
if x == 0:
    ggT = y * (2 ** k)
else:
    ggT = x * (2 ** k)

print("ggT " + f"{ggT}")
```

# Aufgabe 3

```python
import random

notes = []

for students in range(100):
    note1 = random.randrange(1, 7)
    notes.append(note1)
    notes.sort()    
    notes_total = sum(notes)
    
    arith_mittel = float((notes_total) / 100) 
    
    erw = notes_total / len(notes)
    erw2 = sum(item ** 2 for item in notes) / len(notes)
    var = erw2 - (erw ** 2)
    stdev = var ** 0.5
    
print(notes)
print(f"Das Minimum ist:", notes[0])
print(f"Das Maximum ist:", notes[-1])
print(f"Das arithmetische Mittel ist: {arith_mittel}")
print(f"Standardabweichung ist: {stdev}")
```

# Aufgabe 4

```python
#Eingabe aller Variablen durch den Benutzer

import random

n = int(input ("Bitte geben Sie ein wie viele Nächte Sie gespielt haben "))
k = int(input ("Bitte geben Sie ein wie viele Spiele Sie pro Nacht gespielt haben "))
e = int(input ("Bitte geben Sie wie viel Euro Sie pro Spiel gesetzt haben "))
#g = Gewinn, wird hier als 0 definiert, weil er 0 ist befor angefangen wird zu spielen
g = 0

#x = Wie viele Spiele wurden insgesamt gespielt, so lange läuft die While-Schliefe
x = n * k
counter = x
while counter > 0:
    

    #colorplayed = Farbe auf die zufällig gesetzt wird
    colorplayed = ('black','red')
    def selectRandom(colorplayed):
        return random.choice(colorplayed)

    #colorwon = Farbe auf der die Kugel landet
    colorwon = ('black','red')
    def selectRandom(colorwon):
        return random.choice(colorwon)
    
    #gesetzte Farbe wird mit Farbe, die gewonnen hat verglichen
    if selectRandom(colorplayed) == selectRandom(colorwon):
        #y = Gewinn in einer Runde
        y = (e * 2)
    g += y
    counter -= 1

    
print ("Der Gewinn beträgt", g , "Euro")
```
