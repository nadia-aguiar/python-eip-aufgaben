```python
#a)
def is_leap_year(year):
    '''Nimmmt ein Jahr entgegen und gibt Boolean zurück (True, falls ein Schaltjahr vorliegt; False sonst)'''
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False
    
#b)
def translate_week_day(day):
    '''Nimmt einen Tagesindex entgegen, und gibt den Wochentag zurück'''
    wochentag = ["Sun","Mon","Tue","Wed","Thu","Fri","Sat"]
    if day < len(wochentag):
        return wochentag[day]
    else:
        print("No day.Invalid Argument")
        
#c) 
def get_days_per_month(month):
    '''Nimmt einen Monat entgegen und gibt dessen Anzahl an Tagen zurück (Schaltjahre nicht inbegriffen)'''
    day = int()
    if month == (1 or 3 or 5 or 7 or 8 or 10 or 12):
        day=31
    elif month == 2:
        day = 28
    else:
        day = 30
    return day
    
#d)
def get_week_day(year, month, day):
    '''Nimmt Datum (Jahr, Monat und Tag) entgegen und gibt den Wochentag zurück, auf den das angegebene Datum fällt'''
    a = int((14 - month) / 12)
    y = year - a
    m = int(month + (12 * a) - 2)
    d = int(day + y + int(y/4) - int(y/100) + int(y/400)+((31*m) / 12)) % 7
    return d

#e)
def get_month_calender(y, m):
    '''Nimmt Jahr und Monat entgegen und gibt ein Dictionary zurück, 
    in dem alle Tage des gleichen Wochentags den gleichen Index haben (0 Sun, 1 Mon, 2 Tue, ...)'''
    first_day_of_month_weekday_index = get_week_day(y, m, 1)
    # Generiere Tabelle mit m Zeilen und n Spalten - m=7, n=5
    m = 7
    n = 5
    table = [[0] * n for i in range(m)]
    # Tabelle liegt vor in Tabelle[Zeile][Spalte]. Wir haben 7 Zeilen und 5 Spalten
    if (m == 2 and is_leap_year(y)):
        max = 29
    else:
        max = get_days_per_month(m)
    i = 1
    column = 0
    row = 0    
    while i <= max:
        if i % 7 == 0:
            column += 1
        if i == 1:
            column = 0
            row = first_day_of_month_weekday_index
        else:
            row = row % 7
        table[row][column] = i
        row += 1
        i += 1
    
    list_zero = []
    list_one = []
    list_two = []
    list_three = []
    list_four = []
    list_five = []
    list_six = []

    for j in range(5):
        if(table[0][j] != 0):
            list_zero.append(table[0][j])
        if(table[1][j] != 0):
            list_one.append(table[1][j])
        if(table[2][j] != 0):
            list_two.append(table[2][j])
        if(table[3][j] != 0):
            list_three.append(table[3][j])
        if(table[4][j] != 0):
            list_four.append(table[4][j])
        if(table[5][j] != 0):
            list_five.append(table[5][j])
        if(table[6][j] != 0):
            list_six.append(table[6][j])
        
    a = {
        0: list_zero,
        1: list_one,
        2: list_two,
        3: list_three,
        4: list_four,
        5: list_five,
        6: list_six
    }

    return a

    
    
#f)
def print_month_calender(year, month):
    '''Nimmt Jahr und Monat entgegen und gibt den entsprechenden Monatskalender aus'''
    for i in range(7):
        print((translate_week_day(i)) +  " : " + str((get_month_calender(year, month)[i])))    

#g)
monthdict = {
    1: "January",
    2: "February",
    3: "March",
    4: "April",
    5: "May",
    6: "June",
    7: "July",
    8: "August",
    9: "September",
    10: "October",
    11: "November",
    12: "December"
    
}

def print_calender(year):
    '''Nimmt ein Jahr entgegen und gibt (1) den Monatskalender des Dezembers des Vorjahres, 
    (2) alle Monatskalender des angegebenen Jahres und (3) den Monatskalender des Januars des nächsten Jahres
    aus'''
    print(monthdict[12] + " " + str(year - 1))
    print_month_calender(year - 1, 12)
    i = 1
    print("")
    while i <= 12:
        print(monthdict[i] + " " + str(year))
        print_month_calender(year, i)
        print("")
        i += 1
    print(monthdict[1] + " " + str(year + 1))
    print_month_calender(year + 1, 1)

    
#day = get_week_day(2022, 1, 6)
#translate_week_day(day) 
#print(get_week_day(2022, 1, 9))
#print(get_month_calender(2021, 12))
#print_month_calender(2022, 1)
   
#translate_week_day(7)
#is_leap_year(2020)
#get_days_per_month(5)
#print_calender(2022)
```
