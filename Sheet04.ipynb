3. Gib a aus.
4. Streiche a von der Liste.
5. Gehe nun die Liste durch.
    1. Gehe zur nächsten Zahl in der Liste, die nicht durchgestrichen ist
    2. Prüfe, ob die aktuelle Zahl ein Vielfaches von a ist.
        1. Falls die aktuelle Zahl ein Vielfaches von a ist (aktuelle_zahl modulo a ist 0), streiche sie von der Liste

### Der Gesamtausdruck
Gegeben ist der Ausdruck $\phi \Leftrightarrow (\neg(a \land b) \land (a \lor b)) \lor ((a \land b) \lor \neg(a \lor b))$.

Sei $c \Leftrightarrow (\neg(a \land b) \land (a \lor b))$ und $d \Leftrightarrow ((a \land b) \lor \neg(a \lor b))$; dann ist $\phi \Leftrightarrow c \lor d$.

### c umschreiben

$c \Leftrightarrow \neg(a \land b) \land (a \lor b) \\
\overset{3.1}{\Leftrightarrow} (\neg(a \land b)\land a) \lor (\neg(a \land b) \land b) \\
\overset{8.1}{\Leftrightarrow} ((\neg a \lor \neg b) \land a) \lor ((\neg a \lor \neg b) \land b) \\
\overset{1.2}{\Leftrightarrow} (a \land (\neg a \lor \neg b)) \lor (b \land (\neg a \lor \neg b)) \\
\overset{3.1}{\Leftrightarrow} ((a \land \neg a) \lor (a \land \neg b)) \lor ((b \land \neg a) \lor (b \land \neg b)) \\
\overset{9.2}{\Leftrightarrow} (0 \lor (a \land \neg b)) \lor ((b \land \neg a) \lor 0) \\
\overset{1.1}{\Leftrightarrow} ((a \land \neg b) \lor 0) \lor ((b \land \neg a) \lor 0) \\
\overset{5.2}{\Leftrightarrow} (a \land \neg b) \lor (b \land \neg a)$


Also: $c \Leftrightarrow (a \land \neg b) \lor (b \land \neg a)$.

### Negation von d umschreiben
$d \Leftrightarrow((a \land b) \lor \neg(a \lor b))$, also ist 

$\neg d \Leftrightarrow \neg((a \land b) \lor \neg (a \lor b)) \\
\overset{8.2}{\Leftrightarrow} \neg(a \land b) \land \neg \neg (a \lor b) \\
\overset{7}{\Leftrightarrow} \neg(a \land b) \land (a \lor b) \\
\overset{8.1}{\Leftrightarrow} (\neg a \lor \neg b) \land (a \lor b) \\
\overset{3.1}{\Leftrightarrow} ((\neg a \lor \neg b) \land a) \lor ((\neg a \lor \neg b) \land b) \\
\overset{1.2}{\Leftrightarrow} ((a \land (\neg a \lor \neg b)) \lor ((b \land \neg a) \lor (b \land \neg b)) \\
\overset{3.1}{\Leftrightarrow} ((a \land \neg a) \lor (a \land \neg b)) \lor ((b \land \neg a) \lor (b \land \neg b)) \\
\overset{2.2}{\Leftrightarrow} (a \land \neg a) \lor (a \land \neg b) \lor (b \land \neg a) \lor (b \land \neg b) \\
\overset{9.2}{\Leftrightarrow} 0 \lor (a \land  \neg b) \lor (b \land \neg a) \lor 0 \\
\overset{1.1}{\Leftrightarrow} (a \land  \neg b) \lor (b \land \neg a) \lor 0 \lor 0 \\
\overset{5.2}{\Leftrightarrow} (a \land  \neg b) \lor (b \land \neg a) $.

Somit $(\neg d) \Leftrightarrow (a \land  \neg b) \lor (b \land \neg a)$


### Daraus folgt...
Somit haben wir sowohl $c \Leftrightarrow (a \land  \neg b) \lor (b \land \neg a)$ als auch $ (\neg d) \Leftrightarrow (a \land  \neg b) \lor (b \land \neg a)$. Daraus folgt, dass $c \Leftrightarrow neg d$.

Somit steht im Gesamtausdruck $\phi \Leftrightarrow c \lor d \Leftrightarrow \neg d \lor d \overset{1.1}{\Leftrightarrow} d \lor \neg d \overset{9.1}{\Leftrightarrow} 1$.

Da $\phi \Leftrightarrow 1$ muss $\phi$ (also der Gesamtausdruck) immer den Wert $\textbf{True}$ annehmen.
